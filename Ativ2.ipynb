3. Crie um único gráfico de dispersão com o GDP para todos os países da América do Sul, juntamente com a média histórica do GDP para os países mencionados, no período 1960-2013; no eixo x do gráfico constará o ano (1960-2013) e o eixo y terá a média do GDP daquele ano, considerando todos os países; o desvio padrão será calculado a partir da média histórica do GDP para todos os países da América do Sul

```python
media=banco_gdp_america_sul.mean() # Média dos anos
banco3=banco_gdp_america_sul.copy().T # cópia do banco transposto
banco3=banco3.drop('Country Name') # Retirando a linha com os Country Name
banco3.columns = banco3.iloc[0] # reajustando o indice das colunas
banco3 = banco3.drop('Country Code') # retirando a linha com os Country Code
banco3['media']=media # acrescentando a média dos anos na coluna
banco3.head() 
```

```python
plt.figure(figsize=(18, 8))
cores = ['black', 'grey', 'silver', 'lightcoral', 'chocolate', 'peru', 'wheat', 'gold', 'yellow', 'lime', 'cyan', 'blue', 'pink' ]
for i, pais in enumerate(banco3.columns):
    plt.scatter(banco3.index, banco3[pais], color=cores[i], label=pais)
plt.xlabel("Anos")
plt.ylabel("GDP (em Dolares Americanos)")
plt.title("GDP na America do sul (1960-2013)")
plt.xticks(rotation=45)
plt.legend(title='')
plt.show()
```

4. Identifique os anos e os países para os quais o GDP estaria distante da média (valores 'NaN' não devem ser considerados): 

4.1. Para cada ano, é calculada uma média dos valores de GDP para cada país;

4.2. A média histórica é calculada com a média dos anos;

4.3. A distância da média é verificada pelo desvio padrão (DP): o limite superior igual a média mais dois DPs e o limite inferior igual a média menos dois DPs.

```python
banco4=banco3.T
banco4.drop('media',axis=0)
```

```python
desvio_padrao=banco4.std() # Desvio padrão por ano
paises_fora_2_desvios_por_ano = []

for ano in banco4.columns:
    # Identificando os países fora de 2 dp em cada ano
    paises_fora_2_desvios = banco4.index[(banco4[ano] > media[ano] + 2 * desvio_padrao[ano]) | (banco4[ano] < media[ano] - 2 * desvio_padrao[ano])].tolist()
    
    # Adicionando os países fora de 2 dp em cada ano
    paises_fora_2_desvios_por_ano.append(paises_fora_2_desvios)
```

```python
# Mostrando os países fora de 2 dp em cada ano
for i, ano in enumerate(banco4.columns):
    if paises_fora_2_desvios_por_ano[i]:
        print(f"Países fora de 2 dp em {ano}: {paises_fora_2_desvios_por_ano[i]}")
```

8. Descreva as estatísticas sobre o indicador GDP para os países da América do Sul, com uma tabela que contenha o número de medidas válidas, o número de medidas ausentes, a média, o mínimo, o máximo, a mediana, o desvio padrão e a variância do GDP por país

```python
num_meds=len(banco3.index)
medidas_invalidas=banco3.iloc[:,0:12].isna().sum(axis=0).values
tabela_descritiva=pd.DataFrame({
    'Country Name': banco_gdp_america_sul['Country Name'],
    'Country Code': banco_gdp_america_sul['Country Code'],
    'medidas validas':num_meds-medidas_invalidas,
    'medidas invalidas':medidas_invalidas,
    'media':banco3.iloc[:,0:12].mean().values,
    'minimo':banco3.iloc[:,0:12].min().values,
    'maximo':banco3.iloc[:,0:12].max().values,
    'mediana':banco3.iloc[:,0:12].median().values,
    'desvio padrao':banco3.iloc[:,0:12].std().values,
    'variancia':banco3.iloc[:,0:12].var().values
})
tabela_descritiva
```
